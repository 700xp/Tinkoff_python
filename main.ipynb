```python
corr_df = df_weeks['количество транзакций'].corr()
tmp = corr_df.where(np.triu(np.ones(corr_df.shape),k = 1).astype(bool)).stack().sort_values(ascending = False)
tmp.rename({'кафе, рестораны':'кафе и рестораны'},inplace = True)
tmp.head(8).plot.bar(ylabel = 'корреляция', xlabel = 'пара категорий', ylim = (0.8,0.92), color = 'maroon', title = 'Топ корреляций', rot = 70, figsize = (12,6))
```

## Гипотезы
По данной диаграмме видны топ-5 максимальных значений корреляции для количества транзакций. По ним можно выдвинуть следующие 3 гипотезы:
1) Покупка авиабилетов коррелируется с походами в кафе и/или рестораны, потому что авиабилеты - это часто отпуск или работа, а значит отсутствие желания или времени самостоятельно готовить. При этом, покупка авиабилетов часто подразумевает финансовую возможность людей поесть в кафе и/или ресторане.
2) Примерно по этой же причине, покупка номеров в отелях коррелируется с покупками в кафе и ресторанах, потому что люди, проживающие в отелях чаще хотят сходить поесть куда-то, нежели готовить еду самостоятельно.
3) Траты на развлечения коррелируются с тратами на туры, поскольку тур часто подразумевает поездку в опредленное место на достаточно длинный промежуток времени для отдыха. Поэтому люди, которые покупают туры, хотят разнообразнее проводить дни в их путешествии, то есть тратятся на развлечения.
